```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")            #clean background style to plots

# Load the CSV
def load_csv():
    try:
        df = pd.read_csv("superstore.csv", encoding="latin-1")
        print("\nSuperstore Dataset Loaded Successfully...!\n")
        return df
    except Exception as e:
        print("Error loading dataset:", e)
        return None

# Preview the dataset
def preview_data(df):
    if df is not None:
        print("\n---------- DATA PREVIEW ----------\n")
        print("Rows    :", df.shape[0])
        print("Columns :", df.shape[1])
        print("\nColumn Names:")
        print(df.columns.tolist())
        print("\nFirst 5 Rows:")
        print(df.head())
    else:
        print("Dataset not loaded...")

# Dataset summary
def dataset_summary(df):
    if df is not None:
        print("\n---------- DATASET INFO ----------\n")
        df.info()           #data types, non-null values, memory usage
        print("\n---------- DESCRIPTIVE STATISTICS ----------\n")
        print(df.describe())        #mean, min, max for numerical columns
    else:
        print("Dataset not loaded...")

# Analyze numeric columns
def analyze_numeric_columns(df):
    print("\n---------- ANALYZE NUMERIC COLUMNS ----------\n")
    numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
    print("Numeric Columns:", numeric_columns)

    col = input("Enter column name: ")

    if col not in numeric_columns:
        print("Invalid numeric column name...!")
        return

    print("\n1.Sum\n 2.Mean\n 3.Min\n 4.Max\n 5.Std\n 6.Count\n 7.All")
    ch = input("Enter your choice: ")

    if ch == "1":
        print("Sum: ", df[col].sum())
    elif ch == "2":
        print("Mean: ", df[col].mean())
    elif ch == "3":
        print("Min:", df[col].min())
    elif ch == "4":
        print("Max: ", df[col].max())
    elif ch == "5":
        print("Standard Deviation: ", df[col].std())
    elif ch == "6":
        print("Count: ", df[col].count())
    elif ch == "7":
        print("Complete Summary : ",df[col].describe())
    else:
        print("Invalid choice...!")

# Missing values
def missing_values(df):
    if df is not None:
        print("\n---------- MISSING VALUES ----------\n")
        print(df.isnull().sum())
    else:
        print("Dataset not loaded...")

# Top & Bottom Records
def top_bottom_records(df):
    numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
    print("Numeric Columns:", numeric_columns)
    col = input("Enter Numeric Column Name: ")
    if col in df.select_dtypes(include=np.number).columns:
        print("\n Top 5 Records: ")
        print(df.nlargest(5,col))
        print("\n Bottom 5 Records: ")
        print(df.nsmallest(5,col))
    else:
        print("Invalid Numeric Column...")

# Visualizations
def visualize_data(df):
    print("\n---------- VISUALIZATION ----------\n")
    print("1. Sales Histogram")
    print("2. Profit Distribution")
    print("3. Category-wise Sales")
    print("4. Sales vs Profit")
    print("5. Boxplot (Outliers)")

    ch = input("Enter your choice(1-5): ")

    if ch == "1":
        plt.hist(df['Sales'], bins=10)
        plt.title("Sales Distribution")
        plt.xlabel("Sales")
        plt.ylabel("Frequency")
        plt.show()

    elif ch == "2":
        sns.histplot(df['Profit'], kde=True)
        plt.title("Profit Distribution")
        plt.show()

    elif ch == "3":
        df.groupby('Category')['Sales'].sum().plot(kind='bar')
        plt.title("Category-wise Sales")
        plt.ylabel("Total Sales")
        plt.show()

    elif ch == "4":
        plt.scatter(df['Sales'], df['Profit'])
        plt.xlabel("Sales")
        plt.ylabel("Profit")
        plt.title("Sales vs Profit")
        plt.show()

    elif ch == "5":
        sns.boxplot(x=df['Sales'])
        plt.title("Sales Boxplot (Outliers)")
        plt.show()

    else:
        print("Invalid choice!")

# Correlation analysis
def correlation_analysis(df):
    print("\n---------- CORRELATION ANALYSIS ----------\n")
    num_df = df.select_dtypes(include=np.number)
    sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.show()


# Main menu
def main():
    print("\n WELCOME TO SUPERSTORE DATA ANALYTICS TOOL \n")
    df = load_csv()
    if df is None:
        return

    while True:
        print("\n========== MAIN MENU ==========")
        print("1. Preview Data")
        print("2. Dataset Summary")
        print("3. Analyze Numeric Columns")
        print("4. Missing Values")
        print("5. Visualize Data")
        print("6. Correlation Analysis")
        print("7. Top & Bottom Records")
        print("8. Exit")

        ch = input("Enter your choice (1-8): ")

        if ch == "1":
            preview_data(df)
        elif ch == "2":
            dataset_summary(df)
        elif ch == "3":
            analyze_numeric_columns(df)
        elif ch == "4":
            missing_values(df)
        elif ch == "5":
            visualize_data(df)
        elif ch == "6":
            correlation_analysis(df)
        elif ch == "7":
            top_bottom_records(df)
        elif ch == "8":
            print("\nThank you! Program ended.")
            break
        else:
            print("Invalid choice!")

# Run program
main()
```
